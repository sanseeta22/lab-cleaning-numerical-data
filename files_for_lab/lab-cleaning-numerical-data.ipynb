```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Rename columns to snake case
customer_df.columns = customer_df.columns.str.lower().str.replace(' ', '_')
customer_df
```

```python
# Change 'effective_to_date' column to datetime format
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
customer_df
```

```python
# Create a new dataframe using only numerical data
numerical_df = customer_df.select_dtypes(include=[np.number])
numerical_df
```

```python
# Define a function to differentiate between continuous and discrete variables
def var_type(df):
    continuous_vars = []
    discrete_vars = []
    for col in df.columns:
        if len(df[col].unique()) > 100:
            continuous_vars.append(col)
        else:
            discrete_vars.append(col)
    return continuous_vars, discrete_vars
```

```python
# Separate continuous and discrete variables
continuous_vars, discrete_vars = var_type(numerical_df)
continuous_df = numerical_df[continuous_vars]
discrete_df = numerical_df[discrete_vars]
display(continuous_df)
display(discrete_df)
```

```python
# Plot a correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_df.corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()

#here we can see monthly_premium_auto is highly correlated
#months_since_last_claim,months_since_policy_inception,number_of_open_complaints,number_of_policies are not correlated
```

```python
# Create a function to plot every discrete variable

def plot_discrete_vars(df):
    for col in df.columns:
        plt.figure(figsize=(8, 6))
        sns.countplot(x=col, data=df)
        plt.show()

# Plot every discrete variable
plot_discrete_vars(discrete_df)
```

```python
# Plot every continuous variable # we observes it is right skewed
for col in continuous_df.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(continuous_df[col], bins=20, kde=True)
    plt.show()
```

```python
# Look for outliers in the continuous variables
for col in continuous_df.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=continuous_df[col])
```

```python
# Remove outliers from continuous variables
for col in continuous_df.columns:
    q1 = continuous_df[col].quantile(0.25)
    q3 = continuous_df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    continuous_df = continuous_df[(continuous_df[col] >= lower_bound) & (continuous_df[col] <= upper_bound)]
```

```python
#removing them from your main dataframe.
updated_columns = ['customer_lifetime_value', 'income', 'monthly_premium_auto','total_claim_amount'] 
for col in updated_columns:
    numerical_df[col].update(continuous_df[col])
```

```python
# droping customer column
categorical_df= categorical_df.drop(['customer'], axis = 1)
```

```python
categorical_df['vehicle_class'].value_counts() # we get 6 unique values. we can reduce it to 2
```

```python
#here i am trying to group by the vehicle class
def clean_vehicle_class(x):
  
    if x in ['FourDoor Car', 'TwoDoor Car', 'Sports Car']:
        return 'Sports'
    elif x in ['SUV', 'Luxury SUV','Luxury Car']:
        return 'Luxury'
    else:
        return 'Other'

categorical_df['vehicle_class'] = categorical_df['vehicle_class'].apply(clean_vehicle_class)
```

```python
categorical_df
```

```python
categorical_df['employmentstatus'].value_counts() #we get 5 unique values. We can reduce it as well
```

```python
#function to clean employment status
def clean_employment_status(x):
    if x in ['Employed', 'Unemployed']:
        return 'Employed'
    elif x in ['Medical Leave', 'Disabled', 'Retired']:
        return 'Non-Employment'
    else:
        return 'Other'

categorical_df['employmentstatus'] = categorical_df['employmentstatus'].apply(clean_employment_status)
```

```python
categorical_df
```

```python
categorical_df['education'].value_counts() #we get 5 unique values. We can reduce it as well
```

```python
#function to clean education
def clean_education(x):
    if x in ['Bachelor', 'College']:
        return 'Undergraduate'
    elif x in ['High School or Below']:
        return 'High School or Below'
    elif x in ['Master', 'Doctor']:
        return 'Postgraduate'
    else:
        return 'Other'

categorical_df['education'] = categorical_df['education'].apply(clean_education)
categorical_df
```
